```python
import sklearn.datasets as dataset
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
X, Y = dataset.load_breast_cancer().data, dataset.load_breast_cancer().target
```

```python
proj2D = TSNE(n_components=2, learning_rate='auto',init='random', perplexity=100).fit_transform(X)
proj3D = TSNE(n_components=3, learning_rate='auto',init='random', perplexity=8).fit_transform(X)
```

```python
fig_2d = px.scatter(
    proj2D, x=0, y=1,
    color=Y, labels={'color': 'species'}
)
fig_3d = px.scatter_3d(
    proj3D, x=0, y=1, z=2,
    color=Y, labels={'color': 'species'}
)
fig_3d.update_traces(marker_size=5)

fig_2d.show()
fig_3d.show()
```
